## Importando o cliente

```python
from core.ckan_api_reader import Ckan
import pandas as pd
import matplotlib.pyplot as plt
```

```python
CKAN_DOMAIN = "https://repositorio.seade.gov.br/api/3/"
```

```python
ckan = Ckan(CKAN_DOMAIN)
```

```python
ckan.pkgs
```

## Fazendo download dos dados

```python
df = ckan('microdados-casamentos', search_string='2021', attr='name', how='contains', as_list=True, extract=True, parser_params={'sep': ';'})
```

```python
df = df[0]
```

```python
df
```

```python
df.to_csv('arquivo.csv', sep=';', index=False)
```

## Limpando os dados fornecidos

```python
df
```

```python
cod_reg_sp = 3550308
```

```python
df = df.filter(['naturalcj1','naturalcj2','codrescj1', 'codrescj2', 'idadecj1', 'idadecj2', 'codmunreg'], axis=1)
```

```python
filtro = (df['codrescj1'] == cod_reg_sp) | (df['codrescj2'] == cod_reg_sp)
```

```python
df = df[filtro]
```

```python
df
```

```python
df.reset_index(drop=True)
```

```python
df.head(10)
```

```python
df.columns =['naturalidade_cja', 'naturalidade_cjb', 'codmunicipio_cja', 'codmunicipio_cjb', 'idade_cja', 'idade_cjb', 'codmuncasamento']
```

```python
df.head(10)
```

```python
casamentos_mesmo_cod = (df['codmunicipio_cja'] == cod_reg_sp) & (df['codmunicipio_cjb'] == cod_reg_sp)
```

```python
casamentos_diferentes_cod = (df['codmunicipio_cja'] == cod_reg_sp) | (df['codmunicipio_cjb'] == cod_reg_sp)
```

```python
res = pd.DataFrame({
    'Casamentos com o mesmo código regional': casamentos_mesmo_cod.sum(),
    'Casamentos com códigos regionais diferentes': casamentos_diferentes_cod.sum()
}, index = [''])
```

```python
res.plot(kind='bar', legend=True, title='Casamentos com Códigos Regionais de São Paulo', xlabel = 'Tipo de casamentos', ylabel = 'Numero de casamentos')
plt.legend(["Somente", "Não somente"]);
```
